Monte Carlo simulation can be used to simulate the future price of an underlying asset and an option can then be priced using the simulated value.  The option price itself is not simulated, rather the option price is estimated based on the results of the simulation. This method is most applicable in testing strategies, pricing exotic or long-dated options, and risk management. The method can be further improved by adding consideration for jumps through diffusion.   

To do this, we assume that the time evolution of the price of the asset is a stochastic process similar to a random walk [1] then simulate the random walk. To simulate the random walk, we will use a gaussian log return model.  Resampling is another option but the simulated returns can only reflect what has occurred and thus is not the best representation of what could potentially happen. 

Simulations are run a specified number of times (paths) or until a specified level of precision is reached. Precision is dictated by absolute precision and relative precision parameters, where relative precision represents a percentage of the simulated mean, $rp*\mu$.  In implementation, each simulation ran until either the number of steps is reached or the error in the simulated mean ($\delta \mu$) is less than the maximum of the precision parameters.  As a result, some of the simulations end in convergence (precision parameters satisfied) and others end when the specified number of steps is reached. 

Simulation results are represented as bootstrap algorithm-generated, 90% confidence intervals. The bootstrap algorithm is used because it returns this representation without assuming a specific distribution of the data. This provides an understanding of how much we can trust our simulated estimate, in that we can view the results as there being a 90 percent chance that the “true” mean is between the upper and lower boundary presented. 

## Simulating the Underlying <a class="anchor" id="first-bullet"></a>
A custom simulation class method is developed and called via extension of the MCEngine class [2], shown below. The simulate_many method implements the precision/stopping parameters.

```python
class MCEngine:
    """
    Monte Carlo Engine parent class.
    Runs a simulation many times and computes average and error in average.
    Must be extended to provide the simulate_once method
    """
    def simulate_once(self):
        raise NotImplementedError

    def simulate_many(self, ap=0.1, rp=0.1, ns=1000):
        self.results = []
        s1=s2=0.0
        self.convergence=False
        for k in xrange(1,ns):
            x = self.simulate_once()
            self.results.append(x)
            s1 += x
            s2 += x*x
            mu = float(s1)/k
            variance = float(s2)/k-mu*mu
            dmu = math.sqrt(variance/k)
            if k>10:
                if abs(dmu)<max(ap,abs(mu)*rp):
                    self.convergence = True
                    break
        self.results.sort()
        return bootstrap(self.results, confidence = 0.90)

    def var(self, confidence=95):
        index = int(0.01*len(self.results)*confidence+0.999)
        if len(self.results)-index < 5:
            raise ArithmeticError('not enough data, not reliable')
        return self.results[index]
```

We will use daily adjusted close prices for Microsoft (MSFT) from May 14, 2013 through May 14, 2018 for our simulations. Read the data, get the daily log returns and volatility:

```python
msft = 'C:/Users/Mack/OneDrive/Documents/NW/School/MFE/CSC521monteCarlo/final/msft.CSV'

# read data
msft_data = pd.read_csv(msft)

# convert to date
msft_data['Date'] = pd.to_datetime(msft_data['Date'])
msft_data = msft_data.sort_values('Date') # make sure its in proper order

# get log returns as a field in the df
msft_data['log_returns'] = np.log(msft_data['Adj Close']) - np.log(msft_data['Adj Close'].shift(1))

# get log returns as a series
msft_ln_rets = pd.Series(msft_data['log_returns'][1:]) # get rid of the na return

# get daily mu and sigma
msft_mu = np.mean(msft_ln_rets) 

msft_sigma = np.std(msft_ln_rets) 
```

The gaussian log return model assumes that the log returns of an asset follow a gaussian distribution.  In implementation, python’s random.gauss(mu, sigma) function is used to return a random log return with the probability dictated by the $\mu$ and $\sigma$ arguments passed to the function.  These arguments are found through analysis of historical data, shown above. This process is repeated until a simulated price path is created. An example of the process with 100 simulated price paths for MSFT is presented below.

```python
day = [x for x in xrange(len(msft_data))]
price = pd.Series(msft_data['Adj Close'])

days_to_sim = 252 

plt.figure(1)
plt.ylim(0,250)
plt.xlim(1100,1511)
plt.plot(day, price, color = 'black')
plt.xlabel('Day')
plt.ylabel('MSFT Price')

for scenario in range(100):
    sim_day,sim_price = [],[]
    last_day, P = day[-1], float(price[-1:])
    
    for k in range(days_to_sim):                
        Rt = random.gauss(msft_mu, msft_sigma)
        P = P*exp(Rt)
        sim_day.append(last_day + k)
        sim_price.append(P)
        col = random.choice(['b', 'g', 'r', 'c', 'm', 'y', 'k']) # to get a better color while cycling
        plt.plot(sim_day, sim_price, color = col, linewidth = 0.5)
```

However, the prices within the simulated path are arbitrary as the goal is to estimate a price at a give $t$ in the future.  Subsequently, there is a more efficient implementation of this approach; Consider the following:

$r_t = log\frac{P_t}{P_{t-1}} $

each $r_t$ from historical data gives us $N(\mu_{rt},\sigma_{rt})$

$ e^{r_t} = \frac{P_t}{P_{t-1}}      $ so $P_0$ our last day of price data...\\



$P_1  = P_0e^{r_1}   $ 

we find our first simulated price $P_1$ using our last day of price data and $r_1$ from $N(\mu,\sigma)$

$P_2 = P_1e^{r_2}                    $

$P_3 = P_2e^{r_3} = P_0e^{r_1}e^{r_2}e^{r_3} =   P_0e^{r_1 + r_2 + r_3}$

$P_t = P_0 e ^{N(t * \mu_{rt}, \sqrt t * \sigma_{rt})}$

Because the $\mu$ and $\sigma$ of gaussian distributions scale with time, it is unnecessary to simulate each day during the price path. Only the last historical price, the number of days simulated, and the distribution parameters are needed to simulate the price on the desired day. This efficiency is implemented in the extension of the MCEngine class below.

```python
class Gauss_Underlying(MCEngine):
    
    def __init__(self, p0, mu, sigma, sim_intervals):
        self.p0 = p0 # last price
        self.mu = mu
        self.sigma = sigma
        self.sim_intervals = float(sim_intervals) # number of intervals to simulate
        
    def simulate_once(self):
        p0 = self.p0
        mu = self.mu
        sigma = self.sigma
        sim_intervals = self.sim_intervals 
        
        r_t = random.gauss(mu * sim_intervals,
                           sigma * math.sqrt(sim_intervals))
        
        p_t = p0 * math.exp(r_t)
        
        return p_t
```

Let's look at the class in action by simulating the price of the last trading day of each week in March, April and May from the data.  For each trading week, we will simulate the price on the last trading day of the week based on the price from the first trading day of the week and all of the data up to that day.

The following chart will compare the results of our simulation to the actual prices in the data. Red error bars are used to to represent the 90% confidence interval for each simulated price estimate, the black line is used to show actual prices.

```python
msft_mar_may_18 = msft_data[(msft_data['Date'] >= '2018-03-01')]


date_simulated,estimate,lower, upper = [],[],[],[]

for row in msft_mar_may_18.itertuples():
    row_index = row[0]
    row_dt = row[1] # Date
    row_day = row_dt.weekday()
    # sequence to look at the proper trading week
    # accomodate for short weeks etc
    previous_row_day = msft_data.iloc[row_index - 1][0].weekday()
    
    # for the beginning of the df
    if row_index == 1208:
        week_start = row_dt
        # get simulation data up to and including this day
        msft_p0 = row[6] 
        msft_ln_returns =  pd.Series(msft_data['log_returns'][1:row_index + 1]) 
        msft_mu = np.mean(msft_ln_returns)
        msft_sigma = np.std(msft_ln_returns)
        
    # for the end of the df
    if row_index == 1259:
        break
    
    next_row_day = msft_data.iloc[row_index + 1][0].weekday()
    
    if row_day < previous_row_day:
        week_start = row_dt
        # get simulation data up to and including this day
        msft_p0 = row[6] 
        msft_ln_returns =  pd.Series(msft_data['log_returns'][1:row_index + 1]) 
        msft_mu = np.mean(msft_ln_returns)
        msft_sigma = np.std(msft_ln_returns)
    
    if row_day > next_row_day:
        week_end = row_dt    
        trading_week_duration = float((week_start - week_end).days)
        if trading_week_duration < 0:
            date_simulated.append(pd.to_datetime(week_end)) # the date that we are simulating to
            # simulation sequence
            days_to_sim = -1.0*trading_week_duration
            g = Gauss_Underlying(msft_p0, msft_mu, msft_sigma, days_to_sim)
            result = g.simulate_many(ap = 0.01, rp = .01, ns = 100000)
            lower.append(result[1] - result[0])
            upper.append(result[2] - result[1])
            estimate.append(result[1])

errors = [lower, upper]

# chart the results
fig = plt.figure()
plt.plot(msft_mar_may_18['Date'],msft_mar_may_18['Adj Close'],color = 'black')   
ax = fig.add_subplot(111)
ax.errorbar(date_simulated, estimate, yerr = errors, fmt = 'ro')    
plt.xticks(rotation=90) 
plt.xticks(date_simulated)
plt.title('Weekly Simulations March, April, May')
plt.ylabel('MSFT Price');
```

## Pricing the Option  <a class="anchor" id="second-bullet"></a>
We can further extend our class to price options based on the simulated underlying.  To do this, an option is priced for each simulated path, and the resulting estimated price is given using the bootstrapped confidence interval from the paths.

```python
class Gauss_Option(MCEngine):
    def __init__(self, p0, K, mu, sigma, sim_intervals, annual_rfr):
        self.p0 = p0 # spot price
        self.K = K # strike price
        self.mu = mu
        self.sigma = sigma
        self.sim_intervals = float(sim_intervals) # number of intervals to simulate, intervals to expiry
        self.rfr = annual_rfr # annual risk free rate, used for pv
        
    def simulate_once(self):
        p0 = self.p0
        mu = self.mu
        sigma = self.sigma
        sim_intervals = self.sim_intervals 
        
        r_t = random.gauss(mu * sim_intervals,
                               sigma * math.sqrt(sim_intervals))
        
        p_t = p0 * math.exp(r_t)
        
        return self.pv(self.payoff_fx(p_t))
        
    # option payoffs
    def call_payoff(self, p_t):
        return max(p_t - self.K, 0)
        
    def put_payoff(self, p_t):
        return max(self.K - p_t, 0)
        
    # present value
    def pv(self, payoff_fx):
        daily_rfr = self.rfr / 252 # if days arent being used, this must be adjusted
        return payoff_fx * math.exp(-daily_rfr * self.sim_intervals)
        
```

If we assume that the price of MSFT is $\$96$ and we want to price a call option that expires in 20 days with a strike price of $\$90$:


```python
x = Gauss_Option(96, 90, msft_mu, msft_sigma, 20, .05)
x.payoff_fx = x.call_payoff # set the proper payoff for a call option
x.simulate_many(ap = 0.01, rp = .01, ns = 100000) 
```

The point estimate is $\$8.23$ which indicates that the simulated price in 20 days is roughly $\$98$

For a put option with a strike of $\$100$ that expires in 20 days:

```python
x = Gauss_Option(96, 100, msft_mu, msft_sigma, 20, .05)
x.payoff_fx = x.put_payoff # set the proper payoff for a put option
x.simulate_many(ap = 0.01, rp = .01, ns = 100000) 
```
